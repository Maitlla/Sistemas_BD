**Mapas con folium e dataframes**

```python
# Carga datos de marcas desde un ficheiro e píntaas sobre un mapa

# puntos de interese: (datasets)puntos_cidade.csv 

# Nota: Lembra que tes que engadir as marcas ao mapa unha a unha.
```

```python
import pandas as pd
import folium
```

```python
df_restaurantes = pd.read_csv('../datasets/puntos_cidade.csv',sep=';')
df_restaurantes.head()
```

```python
m = folium.Map(location=[43.38, -8.42],zoom_start=11, width=800, height=600)

for  index, restaurante in df_restaurantes.iterrows():
    folium.Marker([restaurante['latitude'],restaurante['lonxitude']], popup=restaurante['nome'], icon=folium.Icon(icon='cutlery',color='purple')).add_to(m)

m
```
