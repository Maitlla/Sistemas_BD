```python
# Que el índice sea de tipo DateTime
df_covid.index = pd.to_datetime(df_covid.index)

df_covid = df_covid.sort_index()

# Filtrar los datos para el mes de enero de 2021
datos_xaneiro = df_covid.loc['2021-01-01':'2021-01-31', ['pais', 'mortes']]

# Filtrar los datos por país
df_spain = datos_xaneiro[datos_xaneiro['pais'] == 'Spain']
df_portugal = datos_xaneiro[datos_xaneiro['pais'] == 'Portugal']
df_france = datos_xaneiro[datos_xaneiro['pais'] == 'France']

# Configurar el tamaño de la figura
plt.figure(figsize=(20, 7))

# Establecer etiquetas de los ejes
plt.xlabel('Fecha')
plt.ylabel('Número de muertes por día')

# Graficar las series temporales para cada país
sns.lineplot(data=df_spain['mortes'])
sns.lineplot(data=df_portugal['mortes'])
sns.lineplot(data=df_france['mortes'])

plt.legend(['España', 'Portugal', 'Francia'])

plt.title('Evolución de las muertes en diferentes países durante enero de 2021')

plt.show()
```

```python
# 3.2 Try to change the color of the lines assign a new color to each country.
# In addition to changing the color increase the width of the lines
# 3.2 Investiga para cambiar los colores de las líneas del gráfico y asignar un nuevo color a cada país
# Además de cambiar el color, aumenta el ancho de las líneas
# https://seaborn.pydata.org/generated/seaborn.lineplot.html
```

```python
# Que el índice sea de tipo DateTime
df_covid.index = pd.to_datetime(df_covid.index)

df_covid = df_covid.sort_index()

# Filtrar los datos para el mes de enero de 2021
datos_xaneiro = df_covid.loc['2021-01-01':'2021-01-31', ['pais', 'mortes']]

# Filtrar los datos por país
df_spain = datos_xaneiro[(datos_xaneiro['pais'] == 'Spain') & (datos_xaneiro.mortes != 0)]
df_portugal = datos_xaneiro[(datos_xaneiro['pais'] == 'Portugal') & (datos_xaneiro.mortes != 0)]
df_france = datos_xaneiro[(datos_xaneiro['pais'] == 'France') & (datos_xaneiro.mortes != 0)]

# Configurar el tamaño de la figura
plt.figure(figsize=(20, 7))

# Establecer etiquetas de los ejes
plt.xlabel('Fecha')
plt.ylabel('Número de muertes por día')

# Graficar las series temporales para cada país
sns.lineplot(data=df_spain['mortes'], color='purple', linewidth=2.5)
sns.lineplot(data=df_portugal['mortes'], color='green', linewidth=2.5)
sns.lineplot(data=df_france['mortes'], color='red', linewidth=2.5)

plt.legend(['España', 'Portugal', 'Francia'])

plt.title('Evolución de las muertes en diferentes países durante enero de 2021')

plt.show()
```
