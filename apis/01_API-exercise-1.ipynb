### Sample APIs

A simple, no fuss, no mess, no auth playground for learning RESTful or GraphQL APIs.

```python
# Configure the endpoint
# Page with sample APIs: https://sampleapis.com/
# Use the endpoint url for la API que sea
# Copia o enderezo do endpoint
endpoint = 'https://api.sampleapis.com/wines/reds'
```

```python
# Make a first request to the endpoint
# fai unha primeira petición contra o endpoint
requests.get(endpoint)
```

```python
# Query the status code
response = requests.get(endpoint)
response.status_code
```

```python
# Check the status of the response
# Verifica o estado da resposta
if response.status_code == 200:
    print('200 OK')
else:
    print('no OK') 
```

```python
# Modify the endpoint and consult the status of the response after a request (force some failure)
# Modifica o endpoint e consulta o estado da resposta tras unha petición (provoca algún fallo)
endpoint = 'https://api.sampleapis.com/winess/reds' # extra character s added
response = requests.get(endpoint)
response.status_code
```

```python
# Check the correct endpoint again and check the raw content of the API response.
# Volve a consultar o endpoint correcto e comproba o contido (raw) da resposta da API
endpoint = 'https://api.sampleapis.com/wines/reds'
response = requests.get(endpoint)
response.content
```

```python
# Visualize the content of the list in JSON format
# visualiza o contido da respoista en formato JSON
response.json()
```

```python
# Check the type of response
# Comproba o tipo da resposta
type(response.json())
```

```python
# Access the first item on the list
# Accede ao primeiro elemento da lista
response.json()[0]
```

```python
# Check the type of the first item in the list
# Comproba o tipo do primeiro elemento da lista
type(response.json()[0])
```

```python
# Consult the "title" of the first element returned by the endpoint
# Consulta o "title" do primeiro elemento que devolve o endpoint
response.json()[0]['winery']
```

```python
# Query the "title" of the penultimate element returned by the endpoint
# Consulta o "title" do penúltimo elemento que devolve o endpoint
response.json()[-2]['winery']
```

```python
# ++ EXTRA
# We can create dataframes with data from a query
# Podemos crear dataframes cos datos dunha consulta
import pandas as pd
df_wine = pd.DataFrame(response.json())
df_wine
```

```python
# ++EXTRA
# We can choose a column as Dataframe index
# Podo escoller unha columna como index do Dataframe
df_wine = df_wine.set_index('id')
df_wine
```
