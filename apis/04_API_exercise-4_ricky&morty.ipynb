```python
# 5. Create a dataframe with ALL the episodes (from all pages)
# 5. Crea un marco de datos con TODOS los episodios (de todas las páginas)
```

```python
url1 = 'https://rickandmortyapi.com/api/episode?page=1'
response1 = requests.get(url1)
data1 = response1.json()['results']

url2 = 'https://rickandmortyapi.com/api/episode?page=2'
response2 = requests.get(url2)
data2 = response2.json()['results']

url3 = 'https://rickandmortyapi.com/api/episode?page=3'
response3 = requests.get(url3)
data3 = response3.json()['results']

df = pd.DataFrame(data1 + data2 + data3)
df
```

```python
# 6. Eliminate the column of the characters (if any) from the dataframe containing all the episodes
# 6. Elimina la columna de los personajes (si corresponde) del dataframe que contiene todos los episodios.
```

```python
url1 = 'https://rickandmortyapi.com/api/episode?page=1'
response1 = requests.get(url1)
data1 = response1.json()['results']

url2 = 'https://rickandmortyapi.com/api/episode?page=2'
response2 = requests.get(url2)
data2 = response2.json()['results']

url3 = 'https://rickandmortyapi.com/api/episode?page=3'
response3 = requests.get(url3)
data3 = response3.json()['results']

df = pd.DataFrame(data1 + data2 + data3)
df.drop(columns='characters')

```
